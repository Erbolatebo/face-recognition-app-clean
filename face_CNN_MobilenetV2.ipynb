# Model CNN

```python
import os
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
from tensorflow.keras.utils import plot_model
import visualkeras

# Параметры
IMG_HEIGHT = 224
IMG_WIDTH = 224
NUM_CLASSES = 4
BATCH_SIZE = 32
EPOCHS = 20
DATA_DIR = "dataset"  # Путь к датасету из save_data (real, fake, photo, mask)

# Функция для загрузки и предобработки данных
def load_and_preprocess_data(data_dir):
    images = []
    labels = []
    label_map = {"real": 0, "fake": 1, "photo": 2, "mask": 3}
    
    for category in label_map.keys():
        category_path = os.path.join(data_dir, category)
        for img_file in os.listdir(category_path):
            img_path = os.path.join(category_path, img_file)
            img = cv2.imread(img_path)
            if img is not None:
                img = cv2.resize(img, (IMG_WIDTH, IMG_HEIGHT))
                img = img / 255.0
                images.append(img)
                labels.append(label_map[category])
    
    images = np.array(images)
    labels = tf.keras.utils.to_categorical(labels, NUM_CLASSES)
    return images, labels

# Создание модели CNN
def create_model():
    model = models.Sequential([
        layers.Input(shape=(IMG_HEIGHT, IMG_WIDTH, 3)),
        layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(256, (3, 3), activation='relu', padding='same'),
        layers.MaxPooling2D((2, 2)),
        layers.Flatten(),
        layers.Dense(512, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(NUM_CLASSES, activation='softmax')
    ])
    
    model.compile(optimizer='adam',
                 loss='categorical_crossentropy',
                 metrics=['accuracy'])
    return model

# Аугментация данных
def create_data_generator():
    datagen = ImageDataGenerator(
        rotation_range=20,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest'
    )
    return datagen

# Функция для построения графиков
def plot_training_history(history):
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Model Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.tight_layout()
    plt.show()

# Основная функция для обучения
def train_model():
    print("Загрузка данных...")
    images, labels = load_and_preprocess_data(DATA_DIR)
    
    X_train, X_val, y_train, y_val = train_test_split(images, labels, test_size=0.2, random_state=42)
    
    model = create_model()
    model.summary()
    
    # Визуализация структуры модели
    plot_model(model, to_file='liveness_model_structure.png', show_shapes=True, show_layer_names=True)
    print("Схема модели сохранена как 'liveness_model_structure.png'")
    
    visualkeras.layered_view(model, to_file='liveness_cnn_visualization.png', legend=True)
    print("Визуализация CNN сохранена как 'liveness_cnn_visualization.png'")
    
    datagen = create_data_generator()
    datagen.fit(X_train)
    
    history = model.fit(
        datagen.flow(X_train, y_train, batch_size=BATCH_SIZE),
        epochs=EPOCHS,
        validation_data=(X_val, y_val)
    )
    
    model.save("liveness_model.h5")
    print("Модель сохранена как 'liveness_model.h5'")
    
    plot_training_history(history)
    
    return model, history

if __name__ == "__main__":
    model, history = train_model()
    print(f"\nРезультаты обучения:")
    print(f"Точность на валидации: {history.history['val_accuracy'][-1]:.4f}")
    print(f"Потери на валидации: {history.history['val_loss'][-1]:.4f}")
```

# Model MonilnetV2

```python
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, BatchNormalization, Dropout
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
from tensorflow.keras.utils import plot_model
from tensorflow.keras.regularizers import l2
from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report
import os
import glob
import numpy as np
import albumentations as A
import logging
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import cv2

# Настройка логирования
logging.basicConfig(filename='accessory_training_log.txt', level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# Параметры
IMG_HEIGHT = 224
IMG_WIDTH = 224
BATCH_SIZE = 32
EPOCHS = 20
ACCESSORY_DATASET_PATH = "accessory_dataset"
ACCESSORY_CLASSES = ['no_accessory', 'glasses', 'headwear', 'medical_mask']
USE_EARLY_STOPPING = True  # False для 20 эпох

# Создание папки для графиков
os.makedirs("plots", exist_ok=True)

# Усиленная аугментация
augmentation = A.Compose([
    A.Rotate(limit=30, p=0.7),
    A.Affine(translate_percent=0.3, scale=(0.7, 1.3), rotate=(-30, 30), shear=(-30, 30), p=0.7),
    A.RandomBrightnessContrast(brightness_limit=0.3, contrast_limit=0.3, p=0.7),
    A.HorizontalFlip(p=0.5),
    A.RandomGamma(gamma_limit=(80, 120), p=0.5),
    A.MotionBlur(blur_limit=5, p=0.4),  # Размытие движения
    A.RandomShadow(shadow_roi=(0, 0.5, 1, 1), num_shadows_limit=(1, 3), p=0.3),  # Тени
    A.GaussNoise(var_limit=(10.0, 50.0), p=0.3),  # Шум
    A.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.2, p=0.5),
])

# Создание синтетических данных
def augment_and_save(image_path, save_dir, num_samples=3):
    os.makedirs(save_dir, exist_ok=True)
    img = cv2.imread(image_path)
    if img is None:
        logging.warning(f"Не удалось загрузить для аугментации: {image_path}")
        return 0
    base_name = os.path.basename(image_path).split('.')[0]
    for i in range(num_samples):
        augmented = augmentation(image=img)['image']
        save_path = os.path.join(save_dir, f"{base_name}_aug_{i}.jpg")
        cv2.imwrite(save_path, augmented)
    return num_samples

# Функция для аугментации изображения
def augment_image(image):
    image_np = image.numpy()
    augmented = augmentation(image=image_np)
    return augmented['image'].astype(np.float32)

# Функция для загрузки изображения
def load_image(path):
    image = tf.io.read_file(path)
    image = tf.image.decode_image(image, channels=3, expand_animations=False)
    image = tf.image.resize(image, [IMG_HEIGHT, IMG_WIDTH])
    image = tf.cast(image, tf.float32) / 255.0
    return image

# Функция для загрузки данных для аксессуаров
def load_accessory_data(path, label, augment=False):
    image = load_image(path)
    if augment:
        image = tf.py_function(augment_image, [image], tf.float32)
        image.set_shape([IMG_HEIGHT, IMG_WIDTH, 3])
    label = tf.one_hot(label, len(ACCESSORY_CLASSES))
    return image, label

# Функция для получения путей и меток
def get_data_paths_and_labels(base_dir, label_map):
    image_paths = []
    labels = []
    for category in label_map.keys():
        class_path = os.path.join(base_dir, category)
        if not os.path.exists(class_path):
            logging.warning(f"Папка не найдена: {class_path}")
            continue
        for filename in glob.glob(os.path.join(class_path, '*.jpg')) + glob.glob(os.path.join(class_path, '*.jpeg')) + glob.glob(os.path.join(class_path, '*.png')):
            try:
                img = cv2.imread(filename)
                if img is not None:
                    image_paths.append(filename)
                    labels.append(label_map[category])
                else:
                    logging.warning(f"Поврежденное изображение: {filename}")
            except Exception as e:
                logging.warning(f"Ошибка чтения {filename}: {e}")
    
    # Балансировка датасета
    class_counts = Counter(labels)
    print("Распределение классов:", class_counts)
    logging.info(f"Распределение классов: {class_counts}")
    avg_count = sum(class_counts.values()) / len(class_counts) if class_counts else 1
    for category, idx in label_map.items():
        count = class_counts.get(idx, 0)
        if count < 0.8 * avg_count:
            class_path = os.path.join(base_dir, category)
            num_needed = int(0.8 * avg_count - count)
            logging.info(f"Добавление {num_needed} синтетических изображений для {category}")
            existing_images = glob.glob(os.path.join(class_path, '*.jpg')) + glob.glob(os.path.join(class_path, '*.jpeg')) + glob.glob(os.path.join(class_path, '*.png'))
            for img_path in existing_images[:num_needed]:
                num_added = augment_and_save(img_path, class_path)
                for i in range(num_added):
                    image_paths.append(os.path.join(class_path, f"{os.path.basename(img_path).split('.')[0]}_aug_{i}.jpg"))
                    labels.append(idx)
    
    print(f"Найдено {len(image_paths)} изображений после балансировки")
    logging.info(f"Найдено {len(image_paths)} изображений после балансировки, распределение классов: {Counter(labels)}")
    return image_paths, labels

# Построение модели для аксессуаров (MobileNetV2)
def build_accessory_model():
    base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(IMG_HEIGHT, IMG_WIDTH, 3))
    for layer in base_model.layers[:-30]:
        layer.trainable = False
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = BatchNormalization()(x)
    x = Dense(128, activation='relu', kernel_regularizer=l2(0.01))(x)  # L2-регуляризация
    x = Dropout(0.6)(x)
    output = Dense(len(ACCESSORY_CLASSES), activation='softmax')(x)
    model = Model(inputs=base_model.input, outputs=output)
    return model

# Функция для построения графиков обучения (аналогично CNN)
def plot_training_history(history):
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history['accuracy'], label='Train Accuracy')
    plt.plot(history['val_accuracy'], label='Validation Accuracy')
    plt.title('Model Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(history['loss'], label='Train Loss')
    plt.plot(history['val_loss'], label='Validation Loss')
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.tight_layout()
    plt.savefig('plots/accessory_training_history.png')
    plt.show()

# Функция для построения матрицы ошибок
def plot_confusion_matrix(y_true, y_pred, classes, title):
    from sklearn.metrics import confusion_matrix
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
    plt.title(title)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.savefig(f'plots/{title.lower().replace(" ", "_")}.png')
    plt.close()

# Функция для построения распределения классов
def plot_class_distribution(labels, classes, title):
    label_counts = np.bincount(labels)
    plt.figure(figsize=(8, 5))
    plt.bar(range(len(classes)), label_counts, tick_label=classes)
    plt.title(title)
    plt.xlabel('Class')
    plt.ylabel('Number of Images')
    plt.savefig(f'plots/{title.lower().replace(" ", "_")}.png')
    plt.close()

# Функция для построения графика ошибок по классам
def plot_class_errors(y_true, y_pred, classes, title):
    errors = [0] * len(classes)
    for t, p in zip(y_true, y_pred):
        if t != p:
            errors[t] += 1
    plt.figure(figsize=(8, 5))
    plt.bar(range(len(classes)), errors, tick_label=classes)
    plt.title(f'{title} Errors')
    plt.xlabel('Class')
    plt.ylabel('Number of Errors')
    plt.savefig(f'plots/{title.lower().replace(" ", "_")}_errors.png')
    plt.close()

# Функция для визуализации архитектуры модели
def plot_model_architecture(model, filename, title):
    plot_model(model, to_file=os.path.join("plots", filename), show_shapes=True, show_layer_names=True)
    print(f"Архитектура модели сохранена в plots/{filename}")
    summary_filename = filename.replace('.png', '_summary.txt')
    with open(os.path.join("plots", summary_filename), 'w', encoding='utf-8') as f:
        model.summary(print_fn=lambda x: f.write(x + '\n'))
    print(f"Текстовое описание архитектуры сохранено в plots/{summary_filename}")

# Функция для отображения аугментированных изображений
def display_augmented_images(image_paths, num_images=10):
    os.makedirs("augmented_accessory", exist_ok=True)
    plt.figure(figsize=(15, 6))
    selected_paths = image_paths[:num_images]
    for i, img_path in enumerate(selected_paths):
        img = cv2.imread(img_path)
        if img is None:
            continue
        aug_img = augmentation(image=img)['image']
        cv2.imwrite(f"augmented_accessory/accessory_aug_{i}.jpg", aug_img)
        plt.subplot(2, 5, i+1)
        plt.imshow(cv2.cvtColor(aug_img, cv2.COLOR_BGR2RGB))
        plt.axis('off')
        plt.title(f'Augmented {i+1}')
    plt.tight_layout()
    plt.show()

# Обучение модели аксессуаров
def train_accessory_model():
    accessory_label_map = {'no_accessory': 0, 'glasses': 1, 'headwear': 2, 'medical_mask': 3}
    accessory_image_paths, accessory_labels = get_data_paths_and_labels(ACCESSORY_DATASET_PATH, accessory_label_map)
    
    if not accessory_image_paths:
        print("Ошибка: Не найдено изображений для аксессуаров")
        logging.error("Не найдено изображений для аксессуаров")
        return None
    
    # Проверка баланса классов
    class_counts = Counter(accessory_labels)
    print("Распределение классов:", class_counts)
    if max(class_counts.values()) / min(class_counts.values()) > 2:
        logging.warning("Несбалансированный датасет. Рекомендуется добавить данные для классов с малым количеством изображений.")
    
    accessory_class_weights = compute_class_weight('balanced', classes=np.unique(accessory_labels), y=accessory_labels)
    accessory_class_weight_dict = dict(enumerate(accessory_class_weights))
    
    train_paths, val_paths, train_labels, val_labels = train_test_split(
        accessory_image_paths, accessory_labels, test_size=0.2, stratify=accessory_labels, random_state=42
    )
    
    train_dataset = tf.data.Dataset.from_tensor_slices((train_paths, train_labels))
    train_dataset = train_dataset.map(
        lambda path, label: load_accessory_data(path, label, augment=True)
    ).shuffle(buffer_size=1000).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
    
    val_dataset = tf.data.Dataset.from_tensor_slices((val_paths, val_labels))
    val_dataset = val_dataset.map(
        lambda path, label: load_accessory_data(path, label, augment=False)
    ).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
    
    accessory_model = build_accessory_model()
    accessory_model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
        loss='categorical_crossentropy',
        metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall()]
    )
    
    callbacks = [ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=1e-6)]
    if USE_EARLY_STOPPING:
        callbacks.append(EarlyStopping(monitor='val_loss', patience=7, restore_best_weights=True))
    
    accessory_history = accessory_model.fit(
        train_dataset,
        validation_data=val_dataset,
        epochs=EPOCHS,
        class_weight=accessory_class_weight_dict,
        callbacks=callbacks,
        verbose=1
    )
    
    accessory_model.save("accessory_model.h5")
    logging.info("Модель аксессуаров сохранена")
    joblib.dump(accessory_history.history, 'accessory_history.pkl')
    
    # Предсказания для анализа
    y_true = []
    y_pred = []
    for images, labels in val_dataset:
        preds = accessory_model.predict(images, verbose=0)
        y_true.extend(np.argmax(labels.numpy(), axis=1))
        y_pred.extend(np.argmax(preds, axis=1))
    
    joblib.dump({'y_true': y_true, 'y_pred': y_pred}, 'accessory_predictions.pkl')
    
    # Вывод classification report
    print("Classification Report for Accessory Model:")
    report = classification_report(y_true, y_pred, target_names=ACCESSORY_CLASSES, zero_division=0)
    print(report)
    logging.info(f"Classification Report:\n{report}")
    
    # Визуализация
    plot_training_history(accessory_history.history)
    plot_confusion_matrix(y_true, y_pred, ACCESSORY_CLASSES, 'Accessory Confusion Matrix')
    plot_class_distribution(accessory_labels, ACCESSORY_CLASSES, 'Accessory Class Distribution')
    plot_class_errors(y_true, y_pred, ACCESSORY_CLASSES, 'Accessory Class')
    plot_model_architecture(accessory_model, 'accessory_model_architecture.png', 'Accessory Model')
    
    print("Отображение аугментированных изображений...")
    display_augmented_images(train_paths, num_images=10)
    
    return accessory_model

if __name__ == "__main__":
    train_accessory_model()
```

```python
import cv2
import numpy as np
import tensorflow as tf
import mediapipe as mp
import logging
import os
from PIL import Image, ImageDraw, ImageFont
from collections import deque
from collections import Counter
import random

# Настройка логирования
logging.basicConfig(filename='testing_log.txt', level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# Параметры
IMG_HEIGHT = 224
IMG_WIDTH = 224
LIVENESS_CLASSES = ["Real", "Fake", "Photo", "Mask"]
ACCESSORY_CLASSES = ["No Accessory", "Glasses", "Headwear", "Medical Mask"]
SLIDING_WINDOW_SIZE = 30
CONFIDENCE_THRESHOLDS = {
    0: 0.6,  # No Accessory
    1: 0.4,  # Glasses
    2: 0.6,  # Headwear
    3: 0.4   # Medical Mask
}
BLINK_THRESHOLD = 0.2
HEAD_MOVEMENT_THRESHOLD = 15
SMILE_THRESHOLD = 3
METHOD_DURATION = 3

# Инициализация Mediapipe
mp_face_detection = mp.solutions.face_detection
mp_face_mesh = mp.solutions.face_mesh
face_detection = mp_face_detection.FaceDetection(min_detection_confidence=0.7)
face_mesh = mp_face_mesh.FaceMesh(max_num_faces=1, refine_landmarks=True)

# Загрузка моделей
try:
    liveness_model = tf.keras.models.load_model("liveness_model.h5")
    logging.info("Liveness model loaded successfully")
except Exception as e:
    print(f"Error loading liveness model: {e}")
    logging.error(f"Error loading liveness model: {e}")
    exit()

try:
    accessory_model = tf.keras.models.load_model("accessory_model.h5")
    logging.info("Accessory model loaded successfully")
except Exception as e:
    print(f"Error loading accessory model: {e}")
    logging.error(f"Error loading accessory model: {e}")
    exit()

# Наложение текста с поддержкой UTF-8
def put_text_pil(frame, text, position, color, font_size=24):
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    pil_img = Image.fromarray(frame_rgb)
    draw = ImageDraw.Draw(pil_img)
    try:
        font = ImageFont.truetype("arial.ttf", font_size)
    except:
        font = ImageFont.load_default()
    draw.text(position, text, font=font, fill=color)
    frame_bgr = cv2.cvtColor(np.array(pil_img), cv2.COLOR_RGB2BGR)
    return frame_bgr

# Предобработка для живучести
def preprocess_frame(frame):
    img = cv2.resize(frame, (IMG_WIDTH, IMG_HEIGHT))
    img = img.astype(np.float32) / 255.0
    img = np.expand_dims(img, axis=0)
    return img

# Предобработка для аксессуаров с CLAHE
def preprocess_image(image):
    img_lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    img_lab[:, :, 0] = clahe.apply(img_lab[:, :, 0])
    img = cv2.cvtColor(img_lab, cv2.COLOR_LAB2BGR)
    img = cv2.resize(img, (IMG_HEIGHT, IMG_WIDTH))
    img = img.astype(np.float32) / 255.0
    img = np.expand_dims(img, axis=0)
    return img

# Подавление шума
def apply_noise_reduction(frame):
    frame = cv2.GaussianBlur(frame, (5, 5), 0)
    return frame

# Сглаживание предсказаний аксессуаров
def smooth_predictions(queue):
    if not queue:
        return None
    avg_probs = np.mean(queue, axis=0)
    accessory_idx = np.argmax(avg_probs)
    confidence = avg_probs[accessory_idx]
    return accessory_idx, confidence, avg_probs.tolist()

# Вычисление EAR
def calculate_ear(eye_landmarks):
    eye_landmarks = np.array(eye_landmarks)
    v1 = np.linalg.norm(eye_landmarks[1] - eye_landmarks[5])
    v2 = np.linalg.norm(eye_landmarks[2] - eye_landmarks[4])
    h = np.linalg.norm(eye_landmarks[0] - eye_landmarks[3])
    ear = (v1 + v2) / (2.0 * h)
    return ear

# Вычисление улыбки
def calculate_smile(mouth_landmarks):
    mouth_landmarks = np.array(mouth_landmarks)
    mouth_width = np.linalg.norm(mouth_landmarks[0] - mouth_landmarks[1])
    mouth_height = np.linalg.norm(mouth_landmarks[2] - mouth_landmarks[3])
    smile_ratio = mouth_width / (mouth_height + 1e-6)
    return smile_ratio

# Проверка моргания
def detect_blink(landmarks, w, h):
    left_eye = [landmarks.landmark[i] for i in [33, 160, 158, 133, 153, 144]]
    right_eye = [landmarks.landmark[i] for i in [362, 385, 387, 263, 373, 380]]
    left_eye_coords = [(lm.x * w, lm.y * h) for lm in left_eye]
    right_eye_coords = [(lm.x * w, lm.y * h) for lm in right_eye]
    left_ear = calculate_ear(left_eye_coords)
    right_ear = calculate_ear(right_eye_coords)
    ear = (left_ear + right_ear) / 2.0
    return ear < BLINK_THRESHOLD, ear

# Проверка улыбки
def detect_smile(landmarks, w, h):
    mouth = [landmarks.landmark[i] for i in [61, 291, 0, 17]]
    mouth_coords = [(lm.x * w, lm.y * h) for lm in mouth]
    smile_ratio = calculate_smile(mouth_coords)
    return smile_ratio > SMILE_THRESHOLD, smile_ratio

# Проверка движения головы
def detect_movement(head_positions):
    if len(head_positions) <= 1:
        return False, "No Data"
    dx = abs(head_positions[-1][0] - head_positions[0][0])
    dy = abs(head_positions[-1][1] - head_positions[0][1])
    moving = dx > HEAD_MOVEMENT_THRESHOLD or dy > HEAD_MOVEMENT_THRESHOLD
    return moving, "Moving" if moving else "Stable"

# Основная функция тестирования
def main():
    cap = cv2.VideoCapture(0)
    if not cap.isOpened():
        print("Error: Could not open camera")
        logging.error("Could not open camera")
        return
    
    print("Camera started. Press 'q' to exit, 'c' to start liveness check.")
    logging.info("Camera started")
    
    accessory_queue = deque(maxlen=SLIDING_WINDOW_SIZE)
    head_positions = deque(maxlen=10)
    liveness_error_count = 0
    accessory_error_count = 0
    accessory_error_stats = Counter()
    blink_detected = False
    smile_detected = False
    movement_detected = False
    random_method = None
    method_timer = 0
    liveness_result = None
    face_detected = False

    while True:
        ret, frame = cap.read()
        if not ret:
            print("Error reading frame")
            logging.error("Error reading frame")
            break
        
        frame = apply_noise_reduction(frame)
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        detection_results = face_detection.process(frame_rgb)
        mesh_results = face_mesh.process(frame_rgb)
        output_frame = frame.copy()
        h, w, _ = frame.shape

        liveness_status = "No Face"
        accessory_status = "No Face"
        liveness_confidence = 0.0
        accessory_confidence = 0.0
        accessory_probs = [0.0] * len(ACCESSORY_CLASSES)
        ear_value = "N/A"
        smile_value = "N/A"
        head_movement = "No Data"

        if not detection_results.detections or not mesh_results.multi_face_landmarks:
            face_detected = False
            output_frame = put_text_pil(output_frame, "No Face Detected", (10, 30), (255, 0, 0), 24)
        else:
            face_detected = True
            output_frame = put_text_pil(output_frame, "Face Detected", (10, 30), (0, 255, 0), 24)

            # Аксессуары: вырезка лица
            bbox = detection_results.detections[0].location_data.relative_bounding_box
            x = int(bbox.xmin * w)
            y = int(bbox.ymin * h)
            width = int(bbox.width * w)
            height = int(bbox.height * h)
            x_offset = int(width * 0.2)
            y_offset = int(height * 0.5)  # Для Headwear
            x = max(0, x - x_offset)
            y = max(0, y - y_offset)
            width = min(w - x, width + 2 * x_offset)
            height = min(h - y, height + y_offset)
            face_img = frame[y:y+height, x:x+width]
            
            # Живучесть: полный кадр
            preprocessed_frame = preprocess_frame(frame)
            liveness_pred = liveness_model.predict(preprocessed_frame, verbose=0)[0]
            liveness_idx = np.argmax(liveness_pred)
            liveness_status = LIVENESS_CLASSES[liveness_idx]
            liveness_confidence = liveness_pred[liveness_idx] * 100

            # Аксессуары: предсказание
            if face_img.size > 0:
                processed_img = preprocess_image(face_img)
                accessory_pred = accessory_model.predict(processed_img, verbose=0)[0]
                accessory_queue.append(accessory_pred)
                smoothed_result = smooth_predictions(accessory_queue)
                if smoothed_result:
                    accessory_idx, accessory_confidence, accessory_probs = smoothed_result
                    if accessory_idx == 0 and accessory_probs[2] > 0.4:  # Headwear
                        accessory_idx = 2
                        accessory_confidence = accessory_probs[2]
                    if accessory_confidence < CONFIDENCE_THRESHOLDS[accessory_idx]:
                        accessory_idx = 0
                        accessory_confidence = accessory_probs[0]
                    accessory_status = ACCESSORY_CLASSES[accessory_idx]
                else:
                    accessory_idx = np.argmax(accessory_pred)
                    accessory_confidence = accessory_pred[accessory_idx]
                    accessory_probs = accessory_pred.tolist()
                    if accessory_idx == 0 and accessory_probs[2] > 0.4:
                        accessory_idx = 2
                        accessory_confidence = accessory_probs[2]
                    if accessory_confidence < CONFIDENCE_THRESHOLDS[accessory_idx]:
                        accessory_idx = 0
                        accessory_confidence = accessory_probs[0]
                    accessory_status = ACCESSORY_CLASSES[accessory_idx]
            else:
                logging.warning("Empty face image for accessory detection")
                accessory_status = "No Face"

            # Анализ активности
            for face_landmarks in mesh_results.multi_face_landmarks:
                nose_tip = face_landmarks.landmark[1]
                nose_x, nose_y = int(nose_tip.x * w), int(nose_tip.y * h)
                head_positions.append((nose_x, nose_y))
                movement_detected, head_movement = detect_movement(head_positions)
                blink_detected, ear = detect_blink(face_landmarks, w, h)
                ear_value = f"{ear:.2f}"
                smile_detected, smile_ratio = detect_smile(face_landmarks, w, h)
                smile_value = f"{smile_ratio:.2f}"

            # Логика живучести
            final_decision = liveness_status
            if liveness_status == "Real":
                if not blink_detected and head_movement == "Stable" and not smile_detected:
                    final_decision = "Suspicious (No Activity)"
                elif random_method == "Blink" and not blink_detected:
                    final_decision = "Fake (No Blinking)"
                elif random_method == "Move" and head_movement == "Stable":
                    final_decision = "Fake (No Movement)"
                elif random_method == "Smile" and not smile_detected:
                    final_decision = "Fake (No Smile)"
            elif liveness_status == "Mask" and (blink_detected or smile_detected):
                final_decision = "Suspicious (Mask Activity?)"

            # Логика случайной проверки
            if method_timer > 0:
                method_timer -= 1 / 30
                if method_timer <= 0:
                    if random_method == "Blink" and blink_detected:
                        liveness_result = "Real"
                    elif random_method == "Move" and movement_detected:
                        liveness_result = "Real"
                    elif random_method == "Smile" and smile_detected:
                        liveness_result = "Real"
                    else:
                        liveness_result = "Fake"
                    random_method = None

            # Подсчёт ошибок (без логирования)
            if final_decision not in ["Real", "Suspicious (No Activity)", "Suspicious (Mask Activity?)"] and (blink_detected or smile_detected or movement_detected):
                liveness_error_count += 1

            if face_img.size > 0 and (accessory_confidence < 0.6 or (accessory_idx == 2 and accessory_confidence < 0.8) or (accessory_idx == 0 and accessory_probs[2] > 0.4)):
                accessory_error_stats[accessory_status] += 1
                accessory_error_count += 1

            # Визуализация
            cv2.rectangle(output_frame, (x, y), (x + width, y + height), (0, 255, 0), 2)
            output_frame = put_text_pil(output_frame, f"Liveness: {final_decision} ({liveness_confidence:.2f}%)", (10, 60), (0, 255, 0), 24)
            output_frame = put_text_pil(output_frame, f"Accessory: {accessory_status} ({accessory_confidence:.2f})", (10, 90), (0, 255, 0), 24)
            for i, (cls, prob) in enumerate(zip(ACCESSORY_CLASSES, accessory_probs)):
                output_frame = put_text_pil(output_frame, f"{cls}: {prob:.2f}", (10, 120 + i * 30), (255, 255, 0), 20)
            output_frame = put_text_pil(output_frame, f"Head: {head_movement}", (10, 240), (255, 255, 0), 20)
            output_frame = put_text_pil(output_frame, f"EAR: {ear_value}", (10, 270), (255, 255, 0), 20)
            output_frame = put_text_pil(output_frame, f"Blink: {'Yes' if blink_detected else 'No'}", (10, 300), (255, 255, 0), 20)
            output_frame = put_text_pil(output_frame, f"Smile: {smile_value} ({'Yes' if smile_detected else 'No'})", (10, 330), (255, 255, 0), 20)
            if random_method:
                output_frame = put_text_pil(output_frame, f"Do: {random_method} ({int(method_timer)}s)", (10, 360), (0, 0, 255), 24)
            if liveness_result:
                color = (0, 255, 0) if liveness_result == "Real" else (0, 0, 255)
                output_frame = put_text_pil(output_frame, f"Result: {liveness_result}", (10, 390), color, 24)

        cv2.imshow("Liveness and Accessory Detection", output_frame)
        
        key = cv2.waitKey(1) & 0xFF
        if key == ord('q'):
            break
        elif key == ord('c') and face_detected:
            random_method = random.choice(["Blink", "Move", "Smile"])
            method_timer = METHOD_DURATION
            liveness_result = None
    
    cap.release()
    cv2.destroyAllWindows()
    
    logging.info(f"Testing completed. Detected {liveness_error_count} liveness errors and {accessory_error_count} accessory errors.")
    logging.info(f"Accessory error statistics: {dict(accessory_error_stats)}")
    print(f"Testing completed. Detected {liveness_error_count} liveness errors and {accessory_error_count} accessory errors.")
    print(f"Accessory error statistics: {dict(accessory_error_stats)}")

if __name__ == "__main__":
    main()
```
