```python
import os
from pathlib import Path
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def show_last_images_in_jupyter(num_images=5):
    base_dir = "accessory_dataset"
    categories = ["no_accessory", "glasses", "headwear", "medical_mask"]
    
    # Проходим по каждой категории
    for category in categories:
        category_path = f"{base_dir}/{category}"
        
        # Проверяем, существует ли директория
        if not os.path.exists(category_path):
            print(f"Директория для категории '{category}' не найдена.")
            continue
        
        # Получаем все файлы изображений в категории
        images = [f for f in os.listdir(category_path) if f.endswith('.jpg')]
        images.sort()  # Сортируем по имени
        
        if not images:
            print(f"В категории '{category}' нет изображений.")
            continue
        
        # Берем последние num_images изображений
        last_images = images[-num_images:]
        
        # Настраиваем отображение
        print(f"\nОтображение последних {len(last_images)} изображений из категории '{category}':")
        plt.figure(figsize=(15, 3))  # Устанавливаем размер фигуры
        
        # Показываем изображения
        for i, img_name in enumerate(last_images, 1):
            img_path = f"{category_path}/{img_name}"
            img = mpimg.imread(img_path)
            
            plt.subplot(1, len(last_images), i)  # Создаем подграфик
            plt.imshow(img)
            plt.title(f"{category} {img_name}")
            plt.axis('off')  # Отключаем оси
        
        plt.show()  # Показываем все изображения для текущей категории

# Вызов функции
show_last_images_in_jupyter(num_images=5)
```
